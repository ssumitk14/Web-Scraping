```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://www.estesparkweather.net/archive_reports.php?date=200901"
req = requests.get(url)
content = req.text

soup = BeautifulSoup(content,'html.parser')
```

### Easy method to extract table from a web page

```python
dfs = pd.read_html(url)
for df in dfs:
    print(df)
```

```python
all = soup.findAll('tr')
counter = 0
d = {'Average temperature':[],
    'Average humidity':[],
    'Average dewpoint':[],
    'Average barometer':[],
    'Average windspeed':[],
    'Average gustspeed':[],
    'Average direction':[],
    'Rainfall for month':[],
    'Rainfall for year':[],
    'Maximum rain per minute':[],
    'Maximum temperature':[],
    'Maximum humidity':[],
    'Maximum pressure':[],
    'Maximum windspeed':[],
    'Maximum heat index':[],
    'Minimum temperature':[],
    'Minimum humidity':[],
    'Minimum pressure':[],
    }
for items in all:
    #print(items)
    try:
        #print(items.findAll('td')[0].text)
        #print(items.findAll('td')[1].text)
        
        avgTemp = items.findAll('td')[0].text
        avgHumidity = items.findAll('td')[0].text
        avgDewpoint = items.findAll('td')[0].text
        avgBaromter = items.findAll('td')[0].text
        avgWindspeed = items.findAll('td')[0].text
        avgGustspeeed = items.findAll('td')[0].text
        avgDirection = items.findAll('td')[0].text
        rainfallMonth = items.findAll('td')[0].text
        rainfallYear = items.findAll('td')[0].text
        maxRainPerMinute = items.findAll('td')[0].text
        maxTemp = items.findAll('td')[0].text
        maxHumidity = items.findAll('td')[0].text
        maxPressure = items.findAll('td')[0].text
        maxWindspeed = items.findAll('td')[0].text
        maxHeatIndex = items.findAll('td')[0].text
        minTemp = items.findAll('td')[0].text
        minHumidity = items.findAll('td')[0].text
        minPressure = items.findAll('td')[0].text
        
        
        
        if "Average temperature" == avgTemp:
            d['Average temperature'].append(items.findAll('td')[1].text)
        if "Average humidity" == avgHumidity:
            d['Average humidity'].append(items.findAll('td')[1].text)
        if "Average dewpoint" == avgDewpoint:
            d['Average dewpoint'].append(items.findAll('td')[1].text)
        if "Average barometer" == avgBaromter:
            d['Average barometer'].append(items.findAll('td')[1].text)
        if "Average windspeed" == avgWindspeed:
            d['Average windspeed'].append(items.findAll('td')[1].text)
        if "Average gustspeed" == avgGustspeeed:
            d['Average gustspeed'].append(items.findAll('td')[1].text)
        if "Average direction" == avgDirection:
            d['Average direction'].append(items.findAll('td')[1].text)
        if "Rainfall for month" == rainfallMonth:
            d['Rainfall for month'].append(items.findAll('td')[1].text)
        if "Rainfall for year" == rainfallYear:
            d['Rainfall for year'].append(items.findAll('td')[1].text)
        if "Maximum rain per minute" == maxRainPerMinute:
            d['Maximum rain per minute'].append(items.findAll('td')[1].text)
        if "Maximum temperature" == maxTemp:
            d['Maximum temperature'].append(items.findAll('td')[1].text)
        if "Maximum humidity" == maxHumidity:
            d['Maximum humidity'].append(items.findAll('td')[1].text)
        if "Maximum pressure" == maxPressure:
            d['Maximum pressure'].append(items.findAll('td')[1].text)
        if "Maximum windspeed" == maxWindspeed:
            d['Maximum windspeed'].append(items.findAll('td')[1].text)
        if "Maximum heat index" == maxHeatIndex:
            d['Maximum heat index'].append(items.findAll('td')[1].text)
        if "Minimum temperature" == minTemp:
            d['Minimum temperature'].append(items.findAll('td')[1].text)
        if "Minimum humidity" == minHumidity:
            d['Minimum humidity'].append(items.findAll('td')[1].text)
        if "Minimum pressure" == minPressure:
            d['Minimum pressure'].append(items.findAll('td')[1].text)
        
        
        #print(items.findAll('td')[1].text)
    except:
        pass
```

```python
df = pd.DataFrame(d)
```
